```python
reddit = praw.Reddit('cyclistNerd') # using auth info from praw.ini
```

```python
MESSAGE_TEXT = \
'''
Hey {username},

I'm cyclistNerd, the moderator of r/photocritique.
I am looking for new members of the subreddit to try out a tool that helps you learn how to write good critiques,
and thought that you might be interested.

We're asking you to use the tool (it's a Google Chrome browser extension) for 10 days to write 10 comments, and
we're raffling off a bunch of nice prizes, including a new camera of your choice! For each comment you write, 
you get an additional raffle ticket.

If you're interested, there are more details [here](https://www.reddit.com/r/photocritique/comments/1opn63v/make_your_critiques_even_better_and_win_a_new/).
Don't hesitate to let me know if you have any questions!
'''

def send_message(username):
    formatted =  MESSAGE_TEXT.format(username=username)
    
    try:
        print(f'Messaging u/{username}...')
        reddit.redditor(username).message(subject='Win a new camera for trying our photocritique tool', message=formatted)
        
    except Exception as e:
        print('Failed.', end='')
        print(e)
        
    finally:
        return
```

```python
for i, username in enumerate( people_to_message[150:200] ):
    print(f'{i:2d} ', end='')
    send_message(username)
    sleep(5*60+ int(random.uniform(-30, 30)))
```
